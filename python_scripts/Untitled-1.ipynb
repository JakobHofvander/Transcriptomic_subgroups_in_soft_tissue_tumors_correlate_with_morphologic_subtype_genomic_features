```python
#### Silpuhette score per diganosis??

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import os
import seaborn as sns

# Set the working directory to a specific path
os.chdir('/path/to/working/dir/')

# 1. Read the gene expression data (CSV file)
df = pd.read_csv('TmV_25_Ano.csv', index_col=0)
#df = df.T
df = pd.DataFrame(df)

### filter these UIDS

print(df)
```

```python
cluster_labels = df.NewDiagnosisFigge_short
genome = df.Genome_group

array = df.drop(columns=['NewDiagnosisFigge_short', 'Genome_group', 'Group']).to_numpy()
print(array)
```

```python
silhouette_scores = silhouette_samples(array, cluster_labels)
print(silhouette_scores)
```

```python
import matplotlib.pyplot as plt

# Example plot_data
plot_data = pd.DataFrame({
    'Silhouette Score': silhouette_scores,
    'Cluster': cluster_labels,
    'Genome': genome
})

# Calculate the mean silhouette score for each cluster and sort by the mean
mean_values = plot_data.groupby('Cluster')['Silhouette Score'].median().sort_values()

# Reorder the clusters by the mean silhouette score
ordered_clusters = mean_values.index

# Create the plot
plt.figure(figsize=(5, 8))
sns.boxplot(y='Cluster', x='Silhouette Score', order=ordered_clusters, data=plot_data, hue='Genome')  # Use the palette based on Genome values

# Overlay a stripplot to show all data points
sns.stripplot(y='Cluster', x='Silhouette Score', data=plot_data, jitter=True, color='black', alpha=0.6, size=2)

# Optional: Add title and labels
plt.title('Silhouette Scores per Sample by Cluster')
plt.xlabel('Silhouette Score')
plt.ylabel('Cluster')

# Add gridlines to the background
plt.grid(True, alpha=0.2)

# Show the plot
plt.savefig('boxplot_Silhouette_per_diagnosis.png', dpi=300, bbox_inches='tight')
plt.show()

print(plot_data)
plot_data.to_csv('Silhouette_per_diagnosis.tsv', sep='\t', index=True)
```
