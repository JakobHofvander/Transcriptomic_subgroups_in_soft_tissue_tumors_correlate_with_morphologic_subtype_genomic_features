```python
#### calculate silhouette score per diagnosis
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import os

# Set the working directory to a specific path
os.chdir('/Users/jhofvand/Dropbox/Jakob+Figge/Transcriptomics/source_data_and_code')

# read the annotated gene expression data file (CSV file)
df = pd.read_csv('TmV_25_Ano.csv', index_col=0)
df = pd.DataFrame(df)
cluster_labels = df.diagnosis
genome = df.Genome_group
# create data array
array = df.drop(columns=['diagnosis', 'Genome_group', 'Group']).to_numpy()
#calculate silhouette scores
silhouette_scores = silhouette_samples(array, cluster_labels)
# save the data 
plot_data = pd.DataFrame({
    'Silhouette Score': silhouette_scores,
    'Cluster': cluster_labels,
    'Genome': genome
})
plot_data.to_csv('Silhouette_per_diagnosis.tsv', sep='\t', index=True)
```
